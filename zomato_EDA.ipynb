```python
cuisines_split =merged_df['Cuisines'].str.split(', ',expand=True)
```

```python
cuisines_split.head()
```

```python
cuisines0=pd.DataFrame(cuisines_split[0].value_counts()).reset_index().rename(columns={0:'0'})
cuisines1=pd.DataFrame(cuisines_split[1].value_counts()).reset_index().rename(columns={0:'1'})
cuisines2=pd.DataFrame(cuisines_split[2].value_counts()).reset_index().rename(columns={0:'2'})
cuisines3=pd.DataFrame(cuisines_split[3].value_counts()).reset_index().rename(columns={0:'3'})
cuisines4=pd.DataFrame(cuisines_split[4].value_counts()).reset_index().rename(columns={0:'4'})
cuisines5=pd.DataFrame(cuisines_split[5].value_counts()).reset_index().rename(columns={0:'5'})
cuisines6=pd.DataFrame(cuisines_split[6].value_counts()).reset_index().rename(columns={0:'6'})
cuisines7=pd.DataFrame(cuisines_split[7].value_counts()).reset_index().rename(columns={0:'7'})
```

```python
merged_cuisines4=pd.merge(merged_cuisines3,cuisines4,how='outer',on='index')
merged_cuisines4
```

```python
merged_cuisines5=pd.merge(merged_cuisines4,cuisines5,how='outer',on='index')
merged_cuisines5
```

```python
merged_cuisines6=pd.merge(merged_cuisines5,cuisines6,how='outer',on='index')
merged_cuisines6
```

```python
merged_cuisines7=pd.merge(merged_cuisines6,cuisines7,how='outer',on='index')
merged_cuisines7
```
